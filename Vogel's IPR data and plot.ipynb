```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pr = int(input("Enter the reservoir pressure : "))
Qo = int(input("Enter the known value of the oil flow rate : "))
pwf = int(input("Enter the known value of flowing bottom hole pressure : "))

Qmax = Qo / ( 1 - (0.2 * (pwf/pr)) - (0.8 * (pwf/pr)**2) )
pwf1 = np.arange(0,pr+1,50)
df = pd.DataFrame(pwf1,columns = ['Pwf'])

def ipr(num):
    return (Qmax *   ( 1 - (0.2 * (num/pr)) - (0.8 * (num/pr)**2) ))

df['Qo']=df['Pwf'].apply(ipr)
display(df)

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(df['Qo'],df['Pwf'])
axes.set_xlabel('Oil Flow Rate (bbl/day)')
axes.set_ylabel('Flowing Bottom Hole Pressure (psi)')
axes.set_xlim(0)
axes.set_ylim(0)
axes.set_title('Vogel\'s Inflow Performance Relationship');
```
