```python
n = int(input("""
If you got Pore Volume and bulk volume press 1
If you got Bulk volume and grain volume press 2
If you got Bulk volume , Dry sample weight and grain density press 3
If you got Length, Diameter of core sample, Dry sample weight and grain density press 4
If you got Dry sample weight, saturated sample weight, density of saturating liquid, weight of saturated sample inside the saturating liquid press 5
"""))


def porosity1():
    Vb = float(input("Enter the Bulk Volume in cubic cm :  "))
    Vp = float(input("Enter the Pore Volume cubic cm :  "))
    phi = Vp/Vb*100
    if 0 < phi <100:
        print("The value of porosity is  {:.2f}%".format(phi) )
    else:
        print("You have given a wrong input")

def porosity2():
    Vb = float(input("Enter the Bulk Volume in cubic cm:  "))
    Vg = float(input("Enter the grain Volume in cubic cm:  "))
    phi = (Vb-Vg)/Vb*100
    if 0< phi <100:
        print("The value of porosity is  {:.2f}%".format(phi) )
    else:
        print("You have given a wrong input")

def porosity3():
    Vb = float(input("Enter the Bulk Volume in cubic cm:  "))
    Wdry = float(input("Enter the Dry sample weight in g : "))
    Gden = float(input("Enter the density of grain in g/cubic cm: "))
    Vg = Wdry / Gden
    phi = (Vb-Vg)/Vb*100
    if 0< phi <100:
        print("The value of porosity is  {:.2f}%".format(phi) )
    else:
        print("You have given a wrong input")

def porosity4():
    Wdry = float(input("Enter the Dry sample weight in g : "))
    Gden = float(input("Enter the density of grain in g/ cubic cm : "))
    L = float(input("Enter the sample length in cm: "))
    D = float(input("Enter the sample diameter in cm: "))
    Vb = (3.14/4)*D**2*L
    Vg = Wdry / Gden
    phi = (Vb-Vg)/Vb*100
    if 0 < phi <100:
        print("The value of porosity is  {:.2f}%".format(phi) )
    else:
        print("You have given a wrong input")



def porosity5():
    Wdry = float(input("Enter the Dry sample weight in g: "))
    Wdryliq = float(input("Enter the saturated sample weight in g: "))
    Lden = float(input("Enter the density of the saturating liquid in g/cubic cm: "))
    Wsat = float(input("Enter the weight of saturated sample inside the saturating liquid in g: "))
    Vb = (Wdryliq - Wsat)/Lden
    Vp = (Wdryliq - Wdry)/Lden
    phi = (Vp/Vb)*100
    if 0 < phi <100:
        print("The value of porosity is  {:.2f}%".format(phi) )
    else:
        print("You have given a wrong input")

if n == 1:
    porosity1()
elif n==2:
    porosity2()
elif n ==3:
    
    porosity3()
elif n==4:
    porosity4()
elif n==5:
    porosity5()
```
